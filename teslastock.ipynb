Question 2 - Extracting Tesla Revenue Data Using Webscraping

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the webpage containing Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Send a GET request to the webpage
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the revenue data
revenue_table = soup.find("table", {"class": "historical_data_table"})

# Initialize lists to store the extracted data
dates = []
revenues = []

# Extract the data from the table rows
for row in revenue_table.find_all("tr"):
    # Extract the table cells
    cells = row.find_all("td")
    if len(cells) == 2:
        # Extract the date and revenue values
        date = cells[0].text
        revenue = cells[1].text

        # Append the values to the respective lists
        dates.append(date)
        revenues.append(revenue)

# Print the extracted revenue data
for date, revenue in zip(dates, revenues):
    print(date, revenue)
```

Question 5 - Tesla Stock and Revenue Dashboard

```python
import matplotlib.pyplot as plt

# Define the ticker symbol for Tesla
ticker = "TSLA"

# Use yfinance to get the Tesla stock data
tesla_data = yf.download(ticker, start="2022-01-01", end="2022-12-31")

# Calculate the rolling mean of the Close prices
tesla_data['MA50'] = tesla_data['Close'].rolling(window=50).mean()
tesla_data['MA200'] = tesla_data['Close'].rolling(window=200).mean()

# Create a line plot of Tesla stock prices
plt.figure(figsize=(12, 6))
plt.plot(tesla_data.index, tesla_data['Close'], label='Close Price')
plt.plot(tesla_data.index, tesla_data['MA50'], label='MA50')
plt.plot(tesla_data.index, tesla_data['MA200'], label='MA200')
plt.title('Tesla Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

Question 6 - GameStop Stock and Revenue Dashboard

```python
# Step 1: Retrieve GameStop stock data using yfinance
gme_data = yf.download("GME", start="2022-01-01", end="2022-12-31")

# Calculate the rolling mean of the Close prices
gme_data['MA50'] = gme_data['Close'].rolling(window=50).mean()
gme_data['MA200'] = gme_data['Close'].rolling(window=200).mean()

# Create a line plot of GameStop stock prices
plt.figure(figsize=(12, 6))
plt.plot(gme_data.index, gme_data['Close'], label='Close Price')
plt.plot(gme_data.index, gme_data['MA50'], label='MA50')
plt.plot(gme_data.index, gme_data['MA200'], label='MA200')
plt.title('GameStop Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Step 2: Scrape GameStop revenue data from the website
revenue_url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

response = requests.get(revenue_url)
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table', class_='historical_data_table')

revenue_data = []
for row in table.find_all('tr')[1:]:
    columns = row.find_all('td')
    year = columns[0].text
    revenue = columns[1].text.replace('$', '').replace(',', '')
    revenue_data.append([year, float(revenue)])

revenue_df = pd.DataFrame(revenue_data, columns=['Year', 'Revenue'])
revenue_df['Revenue'] = revenue_df['Revenue'] / 1000  # Convert revenue to billions

# Create a bar chart of GameStop revenue
fig = go.Figure(data=go.Bar(x=revenue_df['Year'], y=revenue_df['Revenue']))
fig.update_layout(title='GameStop Revenue', xaxis_title='Year', yaxis_title='Revenue (in billions)')
fig.show()
```

```python
import yfinance as yf
from makegraph import Graph

# Define the ticker symbol for Tesla
ticker = "TSLA"

# Use yfinance to get the Tesla stock data
tesla_data = yf.download(ticker, start="2022-01-01", end="2022-12-31")

# Create a graph object
graph = Graph()

# Add a line plot for Close Price
graph.add_line(tesla_data.index, tesla_data['Close'], label='Close Price')

# Add a line plot for 50-day moving average (MA50)
graph.add_line(tesla_data.index, tesla_data['MA50'], label='MA50')

# Add a line plot for 200-day moving average (MA200)
graph.add_line(tesla_data.index, tesla_data['MA200'], label='MA200')

# Set graph title and axis labels
graph.set_title('Tesla Stock Prices')
graph.set_xlabel('Date')
graph.set_ylabel('Price (USD)')

# Display the graph
graph.show()
```
